Import bibliotek

```python
import numpy as np
import pandas as pd
import sklearn

sklearn.__version__
```

Wygenerowanie danych

```python
data = {
    'size': ['XL', 'L', 'M', 'L', 'M'],
    'color': ['red', 'green', 'blue', 'green', 'red'],
    'gender': ['female', 'male', 'male', 'female', 'female'],
    'price': [199.0, 89.0, 99.0, 129.0, 79.0],
    'weight': [500, 450, 300, 380, 410],
    'bought': ['yes','no','yes','no','yes']
}

df_raw = pd.DataFrame(data=data)
df_raw
```

Utworzenie kopii danych

```python
df = df_raw.copy()
df.info()
```

Zmienne typu danych i wstępna eksploracja

```python
for col in ['size', 'color', 'gender', 'bought']:
    df[col] = df[col].astype('category')

df['weight'] = df['weight'].astype('float')

df.info()
```

Wyświetlenie tylko kolumn numerycznych

```python
df.describe()
```

```python
df.describe().T
```

```python
df.describe(include=['category']).T
```

```python
df
```

LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit(df['bought'])
le.transform(df['bought'])
```

Fit + transform

```python
le.fit_transform(df['bought'])
```

```python
le.classes_
```

Przypisanie na stałe

```python
df['bought'] = le.fit_transform(df['bought'])
df
```

JAk chcemy wrócić do pierwotnej postaci

```python
le.inverse_transform(df['bought'])
```

```python
df['bought'] = le.inverse_transform(df['bought'])
df
```

OneHotEncoder

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse=False)
encoder.fit(df[['size']])
```

```python
encoder.transform(df[['size']])
```

```python
encoder.categories_
```

```python
encoder = OneHotEncoder(drop='first',sparse=False)
encoder.fit(df[['size']])
encoder.transform(df[['size']])
```

Standaryzacja - StandardScaler

```python
print(f"{df['price']}\n")
print(f"średnia: {df['price'].mean()}")
print(f"Odchylenie standardowe{df['price'].std():4f}")
```

```python
(df['price'] - df['price'].mean()) / df['price'].std()
```

```python
def standardize(x):
  return(x - x.mean()) / x.std()

standardize(df['price'])
```
